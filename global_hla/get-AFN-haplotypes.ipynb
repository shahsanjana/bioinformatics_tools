```python
import requests
import time
import csv
import pandas as pd
import glob, os, shutil
```

```python
#import from
ids=pd.read_csv("./ids.tsv")
```

```python
source_dir=os. getcwd()
#os.chdir("global_hla_haplotypes")
```

```python
inputs=[]
for index, row in ids.iterrows():
    #http://www.allelefrequencies.net/tools/getrawdata.asp?pop_id=1243&resolved=true
    inputs.append(["http://www.allelefrequencies.net/tools/getrawdata.asp?pop_id="+str(row['id'])+"&resolved=true","AFND"+str(row['id'])+".csv"])
```

```python
def download_url(args):
    t0 = time.time()
    url, fn = args[0], args[1]
    try:
        r = requests.get(url)
        with open(fn, 'wb') as f:
            f.write(r.content)
        return(url, time.time() - t0)
    except Exception as e:
        print('Exception in download_url():', e)
```

```python
t0 = time.time()
for i in inputs:
    result = download_url(i)
    print('url:', result[0], 'time:', result[1])
print('Total time:', time.time() - t0)
```

```python
source_dir
dst = source_dir+"/AFN"
os.mkdir(dst) 
files = glob.iglob(os.path.join(source_dir, "*.csv"))
for file in files:
    if os.path.isfile(file):
        shutil.move(file, dst)
```
